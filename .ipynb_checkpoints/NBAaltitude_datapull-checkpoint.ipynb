```python
lf0 = [[],[],[],[],[],[],[],[]]

for i in seasons:
    for j in activeplayers:
        try:
            dfall = nba_api.stats.endpoints.playerdashboardbyshootingsplits.PlayerDashboardByShootingSplits(season=i,player_id=j).get_data_frames()
            time.sleep(1)
            for k in list(range(0,7)):
                df1 = dfall[k]
                df1['season'] = i
                df1['player'] = j
                lf0[k].append(df1)
                print(k,i,j)
        except:
            print('Failed at:',k,i,j)
            
pdbss0 = pd.concat(lf0[0],axis=0)
pdbss1 = pd.concat(lf0[1],axis=0)
pdbss2 = pd.concat(lf0[2],axis=0)
pdbss3 = pd.concat(lf0[3],axis=0)
pdbss4 = pd.concat(lf0[4],axis=0)
pdbss5 = pd.concat(lf0[5],axis=0)
pdbss6 = pd.concat(lf0[6],axis=0)

```

```python
df = pdbss1.pivot(index=['season','player'],columns='GROUP_VALUE',values=['FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A',
       'FG3_PCT', 'EFG_PCT', 'BLKA', 'PCT_AST_2PM', 'PCT_UAST_2PM',
       'PCT_AST_3PM', 'PCT_UAST_3PM', 'PCT_AST_FGM', 'PCT_UAST_FGM'])

df.columns = [' '.join(col).strip() for col in df.columns.values]

df.to_csv('pdbss1.csv')
```
